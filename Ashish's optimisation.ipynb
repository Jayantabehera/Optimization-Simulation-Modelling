```python
import simpy
import random
#unit m/s^2
accel=0.72
emerge_brake=2.5
opt_brake=0.36

maxspeed=83.3
maxspeed_time=115.7
maxspeed_dist_tr=4820
slow_deceleration_time=231.4
slow_dist_tr=9640

#signal can either be red or green
signal_switch_time=5
block_length=15000
dist_between_train=30000
total_dis=145000

#city speed limit
city_max_speed=31
```

```python
class train:
    def __init__(self,signal=14):
        self.signal=signal
        #self.clock=0
        self.num_train=0
        self.signal_list=[1]*(signal+3)
        self.train_list=[]
    

    def signal_dist(self):
        a=145000/self.signal
        return a
    #as of now just taking random time for POC, after then we will the other variables
    def time_in_block(self):
        return int(random.randint(110,140))
    
        

    
    def moving_forward(self,env,name):
        #print((self.signal_list))
        b=0
        while True:
            
            #if (name==10):
            print(name, now(env))
            if (b==0):
                yield env.timeout((60*5)+5)
                self.signal_list[b]=1
                #self.signal_list[b+1]=0
            elif self.signal_list[b]==1 and self.signal_list[b+1]==1:
                self.signal_list[b]=0
                yield env.timeout(self.time_in_block())
                self.signal_list[b]=1
                
            elif self.signal_list[b]==1 and self.signal_list[b+1]==0:
                self.signal_list[b]=0
                yield env.timeout(self.time_in_block())
                self.signal_list[b]=1
            else:
                yield env.timeout(1)
            if b<len(self.signal_list)-2:
                b=b+1
```

```python
        
train_class=train()
total_train=[]
def daytime(t):
    t=int(t)
    return f"{t//3600:02d}:{(t%3600)//60:02d}:{t%60:02d}"
def now(env):
    return daytime(env.now)
env=simpy.Environment()

def train_proc(env):
    
    i=0
    
    while True:
    #print("train_before " + str(i)+' ', now(env))
        if i != 0:
            temp=train_class.signal_list.index(0)
        else:
            temp=2
        if (temp==2):
            total_train.append(i)
            train_class.signal_list[0]=0
            env.process(train_class.moving_forward(env, i))
        
        yield env.timeout(1)
        i=i+1
        
    #print("train", now(env))
        

env.process(train_proc(env))
env.run(until=3600)
```
